```python
#Import Libraries
import numpy
import sys
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
import joblib
from tkinter import  *
import joblib
from tkinter import messagebox
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

#Import dataset & determine X,y
df = pandas.read_csv("datatest.csv")
features = ['ppm', 'SR', 'repeatofalarm', 'notpokayoke', 'highprice']
X = df[features]
y = df["level"]
#Build a Model & see a Scores
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
k_folds = KFold(n_splits = 5)
scores = cross_val_score(knn, X, y, cv = k_folds)
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
y_true=df
expected=y_true
predict = knn.predict(X)
confusion_matrix(y_true['level'], y_pred=predict)
print(metrics.classification_report(y_true['level'], y_pred=predict))

#Build a joblib for model
joblib.dump(knn, 'my_model_knn.pkl.pkl')
knn_from_joblib = joblib.load('my_model_knn.pkl.pkl')
print(knn_from_joblib.predict(X))

#Build a roots for tkinter
#determine a header,label,frame,
root = Tk()

root.title("Classifier")

root.geometry("800x200")

root.configure(bg="lightblue")

root.config(bg="lightblue")

header = Label(root,text="Classifier",bg="lightblue",
    foreground="black",font=("Arial",15,"bold"))
header.pack()

frame1= Frame(root,bg="lightblue")
frame1.pack()

label1 = Label(frame1,text="PPM",bg="lightblue",foreground="black",
    font=("Arial",15,"bold"))
label1.grid(row=0,column=0,pady=10)

entry1=Entry(frame1,width=10,font=("Arial",15,"bold"),bg="gray",fg="white",
    borderwidth=3)
entry1.grid(row=0,column=1,pady=10)
#########################################################
label2 = Label(frame1,text="S/R",bg="lightblue",foreground="black",
    font=("Arial",15,"bold"))
label2.grid(row=0,column=2,pady=10)

entry2=Entry(frame1,width=10,font=("Arial",15,"bold"),bg="gray",fg="white",
    borderwidth=3)
entry2.grid(row=0,column=3,pady=10)

################################################################
label3 = Label(frame1,text="Repeat of Alarm",bg="lightblue",foreground="black",
    font=("Arial",15,"bold"))
label3.grid(row=0,column=4,pady=10)

entry3=Entry(frame1,width=10,font=("Arial",15,"bold"),bg="gray",fg="white",
    borderwidth=3)
entry3.grid(row=0,column=5,pady=10)
#########################################################
label4 = Label(frame1,text="Not Pokayoke",bg="lightblue",foreground="black",
    font=("Arial",15,"bold"))
label4.grid(row=0,column=6,pady=10)

entry4=Entry(frame1,width=10,font=("Arial",15,"bold"),bg="gray",fg="white",
    borderwidth=3)
entry4.grid(row=0,column=7,pady=10)
###############################################################

label5 = Label(frame1,text="High Price",bg="lightblue",foreground="black",
    font=("Arial",15,"bold"))
label5.grid(row=0,column=8,pady=10)

entry5=Entry(frame1,width=10,font=("Arial",15,"bold"),bg="gray",fg="white",
    borderwidth=3)
entry5.grid(row=0,column=9,pady=10)
###############################################################
level=(Label(root,text=" Please Put a Values in Box ",bg="lightblue",font=("Arial",20,"bold")))
level.pack()

#define predict(classify) for inputs and get a label

def predict():
    ppm = float(entry1.get())
    SR= float(entry2.get())
    repeatofalarm = float(entry3.get())
    notpokayoke = float(entry4.get())
    highprice = float(entry5.get())
    pred=knn_from_joblib.predict([[ppm, SR, repeatofalarm, notpokayoke, highprice]])
    level.config(text="level:"+str([(int(pred))]))

button = Button(text="Grade of Part",bg="green",activebackground="blue",command=predict, borderwidth=3,font=("Arial",20,"bold"))
button.pack()

root.mainloop()
```
